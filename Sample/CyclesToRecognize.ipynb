```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

This is a sample plotting using the "cycles_for_recognition" functions. This is for 1000 particles with two different sample sizes. 

```python
detect_10_1000 = np.genfromtxt("sample_ctr_1000_10.txt", 
                           delimiter = ",", 
                           dtype=float,
                          skip_header=1)
detect_5_1000 = np.genfromtxt("sample_ctr_1000_5.txt", 
                           delimiter = ",", 
                           dtype=float,
                          skip_header=1)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(detect_10_1000[:,2]*100, detect_10_1000[:,1], legend="SP=10")
plt.scatter(detect_5_1000[:,2]*100, detect_5_1000[:,1], label="SP=5")
plt.xlim((0,100))
plt.xlabel("Density Ratio")
plt.ylabel("Cycles")
plt.title("Cycles Needed to Implant a Detectable Memory")
plt.show()
```
